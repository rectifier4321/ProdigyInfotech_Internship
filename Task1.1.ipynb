```python
import pandas as pd
import matplotlib.pyplot as plt

# Correct file path
file_path = 'D:/Desktop/Important Docs/MyCertificates/Prodigy Infotech/Task1/Datav.xlsx'

# Read the data sheet
df = pd.read_excel(file_path)

# Inspect the first few rows to understand the structure
#print(df.head())
print(df.columns)
df1=df.iloc[1:10]

# Ensure 'Country Name' column exists
if 'Country Name' in df1.columns:
    country_col = 'Country Name'
else:
    raise ValueError("Expected column 'Country Name' not found in DataFrame columns")

# Distribution of Categorical Data: Country Name
counts = df1['1960']
country= df1['Country Name']
plt.figure(figsize=(10, 4))
plt.bar(country,counts, color='skyblue')
plt.xlabel('Country Name')
plt.ylabel('Frequency')
plt.title('Distribution of Country Names')
plt.show()

# Distribution of Continuous Data: Population in 1960
population_1960 = df['1960'].dropna()
print(population_1960.describe())

plt.figure(figsize=(8, 4))
plt.hist(population_1960, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Population in 1960')
plt.ylabel('Frequency')
plt.title('Distribution of Population in 1960')
plt.show()

```
